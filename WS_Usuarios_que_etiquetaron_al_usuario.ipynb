```python
#En este caso se utiliza la cuenta de "uadeo.culiacan" como ejemplo para extraer datos
usuario="uadeo.culiacan"#aqui va el usuario al que se le quiere extraer 
driver = uc.Chrome()
#INICIAR INSTAGRAM
driver.get("https://www.instagram.com/")
time.sleep(10)
numa=random.randint(0,100)
username = driver.find_element(by=By.NAME, value="username")
username.click()
username.send_keys("tuusuario")##aqui va el usuario
#na = random.randint(0, numa)
#driver.implicitly_wait(na)
time.sleep(5)
password = driver.find_element(by=By.NAME, value="password")
password.click()
password.send_keys('contraseña')#aqui va la contraseña de tuusuario
#nb= random.randint(0,numa)
time.sleep(5)
login = driver.find_element("css selector", "button[type='submit']").click()
time.sleep(10)
#Buscar el usuario al que se le quiere extraer los datos
searchbox1 = driver.find_element("css selector", "svg[aria-label='Buscar']").click()
time.sleep(3)
searchbox = driver.find_element("css selector", "input[placeholder='Buscar']")
searchbox.clear()
searchbox.send_keys(usuario)
time.sleep(5)
driver.find_element(By.CSS_SELECTOR, ".x1i10hfl:nth-child(1) > .x9f619 > .x9f619 > .x9f619 > .x9f619 > .x9f619 > .x9f619 > .x1lliihq > .x1lliihq").click()
time.sleep(10)
```

```python
#Hacer scroll
time.sleep(10)
scrolldown = driver.execute_script(
    "window.scrollTo(0, document.body.scrollHeight);var scrolldown=document.body.scrollHeight;return scrolldown;")
match = False
while (match == False):
    last_count = scrolldown
    time.sleep(3)
    scrolldown = driver.execute_script(
        "window.scrollTo(0, document.body.scrollHeight);var scrolldown=document.body.scrollHeight;return scrolldown;")
    if last_count == scrolldown:
        match = True

time.sleep(4)
img=[]
vid=[]
links = driver.find_elements("tag name", "a")
for link in links:
    post = link.get_attribute('href') #Encontrar links
    if '/p/' in post:
        img.append(post) #Guardar las imágenes 
    elif 'reel' in post: #Guardar los reels
        vid.append(post)
        if 'reels' in post: #Eliminar los que contienen "reels"
            vid.remove(post)
#Guardar las publicaciones en una variable
publicaciones=vid+img
```

```python
#Abrir las publicaciones una a una
etiquetados=set()
for publicacion in publicaciones:
    print(f"Navigating to: {publicacion}")
    driver.get(publicacion)
    time.sleep(10)
    try:
        boton = driver.find_element(By.CSS_SELECTOR, "button[type='button']")
        if boton:
            boton.click()
            time.sleep(2)
            etiquetas = driver.find_elements(By.XPATH, "//div[contains(@class, '_aa1y')]//a") #Encontrar las etiquetas
            for etiqueta in etiquetas:
                user_link = etiqueta.get_attribute('href')
                if user_link:
                    etiquetados.add(user_link) #Guardar los links de los perfiles de usuario
    except:
        print("No se encontró")
        continue
```

```python
#Limpiar los links dejando lo que esta después de "/" y antes de "/"
etiquetados_porel_usuario=[]
for i, etiquetado in enumerate(etiquetados):
    u = etiquetado.split('/')[-2]
    etiquetados_porel_usuario.append(u)
```
